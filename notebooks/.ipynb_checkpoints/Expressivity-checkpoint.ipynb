# Expressivity of QNN vs Classic NN

## QNN

```python
np.random.seed(42)
backend = Aer.get_backend('qasm_simulator')

layer1 = QLayer(n_qubits=2, n_features=2, n_targets=5, encoder=Encoder(), ansatz=Ansatz(), reps=1, scale=np.pi, backend=backend, shots=100000)
layer2 = QLayer(n_qubits=5, n_features=5, n_targets=5, encoder=Encoder(), ansatz=Ansatz(), reps=1, scale=np.pi, backend=backend, shots=100000)
layer3 = QLayer(n_qubits=2, n_features=2, n_targets=2, encoder=Encoder(), ansatz=Ansatz(), reps=1, scale=1, backend=backend, shots=100000)


layers = [layer1, layer2, layer3]

optimizer = Adam()
network = NeuralNetwork(layers, optimizer)
```

```python
theta = np.linspace(0, 2*np.pi, 1000).reshape(-1,1)
x = (np.cos(theta) + 1)/2
y = (np.sin(theta) + 1)/2
x = np.hstack([x, y])
print(trajectory_length(x))
```

```python
plt.plot(x[:,0], x[:,1])
plt.show()
```

```python
x_trans = network.predict(x)
```

```python
print(trajectory_length(x_trans))
plt.plot(x_trans[:,0], x_trans[:,1])
plt.show()
```
